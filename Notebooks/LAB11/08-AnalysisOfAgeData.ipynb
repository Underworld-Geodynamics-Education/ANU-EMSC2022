## Age / Depth Relationship

We plotted bathymetry and we plotted Ocean age but it is difficult to compare the data because they are on different grids.

I used a global triangulation to resample both datasets at the same points and I made sure that the triangulation had equal areas all over the globe so that each point in the dataset has the same weight as every other **(Why is this not the case in the originals ?)**

This gives us information in the form of a **point dataset** just like the seismicity, and we can plot it accordingly.

```python
# First we import the modules we need

import cartopy
import cartopy.crs as ccrs
import matplotlib
import matplotlib.pyplot as plt

# The universal solvent for arrays (and hence images).
import numpy as np
```

```python
## Read the data

age_depth = np.load("../+Data/OceanAgeDepth.npz")
print(age_depth.files)

lons = age_depth["lons"]
lats = age_depth["lats"]
age  = age_depth["age"]
depth = age_depth["depth"]

```

```python
# Bathymetry

fig = plt.figure(figsize=(12, 12), facecolor="none")
ax  = plt.subplot(111, projection=ccrs.Mollweide())
ax.set_global()

colormap = plt.cm.get_cmap('RdYlBu_r', 10)

m = ax.scatter(lons, lats, 
               c=depth, cmap=colormap, vmin=-6000, vmax=-500.0,
               s=10.0, transform=ccrs.Geodetic())

plt.colorbar(mappable=m, orientation="horizontal", shrink=0.5)
ax.coastlines(resolution="50m", zorder=100, linewidth=0.5)

```

```python
# Age

fig = plt.figure(figsize=(12, 12), facecolor="none")
ax  = plt.subplot(111, projection=ccrs.Mollweide())
ax.set_global()

colormap = plt.cm.get_cmap('RdYlBu', 20)

m = ax.scatter(lons, lats, 
               c=age, cmap=colormap, vmin=0, vmax=200.0,
               s=10.0, transform=ccrs.Geodetic())

plt.colorbar(mappable=m, orientation="horizontal", shrink=0.5)
ax.coastlines(resolution="50m", zorder=100, linewidth=0.5)

```
